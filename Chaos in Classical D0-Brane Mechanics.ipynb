```python
def F(Xi, X):
    
    result = np.zeros([N,N], dtype = complex)
    
    for j in np.arange(9):
        
        result += np.dot(X[j], np.dot(Xi,X[j]) - np.dot(X[j],Xi)) - np.dot(np.dot(Xi,X[j]) - np.dot(X[j],Xi), X[j])
    
    return result
```
