```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
run1 = np.loadtxt("C:\\Users\\cilli\\DIAS.cpp\\Run_files1.0\\run_2.txt", dtype = "double", delimiter = ",")
run1_ds = np.loadtxt("C:\\Users\\cilli\\DIAS.cpp\\Run_files1.0\\run_ds2.txt",dtype = "double",delimiter = ",");
```

```python
sols = np.zeros((8,1000), dtype = "double")
sols_ds = np.zeros((8,1000), dtype = "double")
```

```python
x1 = run1[0:1000]
x2 = run1[1000:2000]
y1 = run1[2000:3000]
y2 = run1[3000:4000]
px1 = run1[4000:5000]
px2 = run1[5000:6000]
py1 = run1[6000:7000]
py2 = run1[7000:8000]

x1_ds = run1_ds[0:1000]
x2_ds = run1_ds[1000:2000]
y1_ds = run1_ds[2000:3000]
y2_ds = run1_ds[3000:4000]
px1_ds = run1_ds[4000:5000]
px2_ds = run1_ds[5000:6000]
py1_ds = run1_ds[6000:7000]
py2_ds = run1_ds[7000:8000]
```

```python
sols = np.array([x1,x2,y1,y2,px1,px2,py1,py2])
sols_ds = np.array([x1_ds,x2_ds,y1_ds,y2_ds,px1_ds,px2_ds,py1_ds,py2_ds])
```

```python
def distance(sols_i, sols_dsi):
    return np.sqrt(
        (sols_i[0] - sols_dsi[0])**2 + (sols_i[1] - sols_dsi[1])**2 + (sols_i[2] - sols_dsi[2])**2 +
        (sols_i[3] - sols_dsi[3])**2 + (sols_i[4] - sols_dsi[4])**2 + (sols_i[5] - sols_dsi[5])**2 +
        (sols_i[6] - sols_dsi[6])**2 + (sols_i[7] - sols_dsi[7])**2
    )
```

```python
korva = distance(sols, sols_ds)
```

```python
times = np.linspace(0,100,1000)
plt.plot(times, np.log(korva))
```
