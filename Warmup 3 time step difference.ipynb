```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
small_step = np.loadtxt("C:\\Users\\cilli\\VS programmes\\smallstep.txt", dtype = "double", delimiter = ",")
big_step = np.loadtxt("C:\\Users\\cilli\\VS programmes\\bigstep.txt", dtype = "double", delimiter = ",")
```

```python
# The coordinates of a system evolved to t = 1000 starting from some thermalised state.
# small step evolves the system with a step size of 10^-4 and big step evolves the system using a step size
# of 1x10^-3

x1_small = small_step[:100]
x2_small = small_step[100:200]
y1_small = small_step[200:300]
y2_small = small_step[300:400]
px1_small = small_step[400:500]
px2_small = small_step[500:600]
py1_small = small_step[600:700]
py2_small = small_step[700:]

x1_big = big_step[:100]
x2_big = big_step[100:200]
y1_big = big_step[200:300]
y2_big = big_step[300:400]
px1_big = big_step[400:500]
px2_big = big_step[500:600]
py1_big = big_step[600:700]
py2_big = big_step[700:]
```

```python
plt.figure(figsize = (15,15))
times = np.linspace(0,1000,100)

plt.subplot(4,2,1)
plt.scatter(times, x1_small - x1_big)
plt.xlabel("time")
plt.ylabel("x1_small - x1_big")

plt.subplot(4,2,2)
plt.scatter(times, x2_small - x2_big)
plt.xlabel("time")
plt.ylabel("x2_small - x2_big")

plt.subplot(4,2,3)
plt.scatter(times, y1_small - y1_big)
plt.xlabel("time")
plt.ylabel("y1_small - y2_big")

plt.subplot(4,2,4)
plt.scatter(times, y2_small - y2_big)
plt.xlabel("time")
plt.ylabel("y2_small - y2_big")

plt.subplot(4,2,5)
plt.scatter(times, px1_small - px1_big)
plt.xlabel("time")
plt.ylabel("px1_small - px1_big")

plt.subplot(4,2,6)
plt.scatter(times, px2_small - px2_big)
plt.xlabel("time")
plt.ylabel("px2_small - px2_big")

plt.subplot(4,2,7)
plt.scatter(times, py1_small - py1_big)
plt.xlabel("time")
plt.ylabel("py1_small - py1_big")

plt.subplot(4,2,8)
plt.scatter(times, py2_small - py2_big)
plt.xlabel("time")
plt.ylabel("py2_small - py2_big");

plt.savefig("asd")
```
